```python
import pandas as pd
import matplotlib.pyplot as plt
import json
import mplcursors

# Load JSON data from file
with open('/content/Atashee Saha.json', 'r') as json_file:
    json_data = json.load(json_file)

# Load Excel data
df = pd.read_csv('/content/Atashee_alpha_beta_ratio.csv')

# Plotting
fig, ax = plt.subplots(figsize=(10, 6))

# Plot Delta
ax.plot(df['time'], df['Delta_filtered'], label='Delta')
# Plot Theta
ax.plot(df['time'], df['Theta_filtered'], label='Theta')
# Plot Alpha1 and Alpha2
ax.plot(df['time'], df['Alpha1_filtered'], label='Alpha1')
ax.plot(df['time'], df['Alpha2_filtered'], label='Alpha2')
# Plot Beta1 and Beta2
ax.plot(df['time'], df['Beta1_filtered'], label='Beta1')
ax.plot(df['time'], df['Beta2_filtered'], label='Beta2')
# Plot Gamma1 and Gamma2
ax.plot(df['time'], df['Gamma1_filtered'], label='Gamma1')
ax.plot(df['time'], df['Gamma2_filtered'], label='Gamma2')

ax.set_xlabel('Time (seconds)')
ax.set_ylabel('Amplitude')
ax.set_title('Brainwave Frequency Bands Over Time')
ax.legend()
ax.grid(True)

# Adding vertical lines for activity spans
activity_spans = [
    ('Reading', json_data['readingStartTime'], json_data['readingTestTime']),
    ('Listening', json_data['listeningStartTime'], json_data['listeningTestTime']),
    ('Video', json_data['videoStartTime'], json_data['videoTestTime'])
]
for span in activity_spans:
    ax.axvline(x=span[1], color='r', linestyle='--', label=span[0] + ' Start')
    ax.axvline(x=span[2], color='g', linestyle='--', label=span[0] + ' End')
    ax.fill_betweenx(y=[min(df.min()), max(df.max())], x1=span[1], x2=span[2], color='yellow', alpha=0.2)
    ax.text((span[1] + span[2]) / 2, max(df.max()), span[0], horizontalalignment='center')

# Make the plot interactive with mplcursors
mplcursors.cursor(hover=True)

plt.tight_layout()

# Save the plot as an image
plt.savefig('interactive_brainwave_frequency_bands_with_json_highlighted_and_labeled.png')

plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import json

# Load JSON data from file
with open('/content/Atashee Saha.json', 'r') as json_file:
    json_data = json.load(json_file)

# Load Excel data
df = pd.read_csv('/content/Atashee_alpha_beta_ratio.csv')

# Plotting
fig, ax = plt.subplots(figsize=(10, 6))

# Plot Delta
ax.plot(df['time'], df['Delta_filtered'], label='Delta')
# Plot Theta
ax.plot(df['time'], df['Theta_filtered'], label='Theta')
# Plot Alpha1 and Alpha2
ax.plot(df['time'], df['Alpha1_filtered'], label='Alpha1')
ax.plot(df['time'], df['Alpha2_filtered'], label='Alpha2')
# Plot Beta1 and Beta2
ax.plot(df['time'], df['Beta1_filtered'], label='Beta1')
ax.plot(df['time'], df['Beta2_filtered'], label='Beta2')
# Plot Gamma1 and Gamma2
ax.plot(df['time'], df['Gamma1_filtered'], label='Gamma1')
ax.plot(df['time'], df['Gamma2_filtered'], label='Gamma2')

ax.set_xlabel('Time (seconds)')
ax.set_ylabel('Amplitude')
ax.set_title('Brainwave Frequency Bands Over Time')
ax.legend()
ax.grid(True)

# Adding vertical lines for activity spans
activity_spans = [
    ('Reading', json_data['readingStartTime'], json_data['readingTestTime']),
    ('Listening', json_data['listeningStartTime'], json_data['listeningTestTime']),
    ('Video', json_data['videoStartTime'], json_data['videoTestTime'])
]
for span in activity_spans:
    ax.axvline(x=span[1], color='r', linestyle='--', label=span[0] + ' Start')
    ax.axvline(x=span[2], color='g', linestyle='--', label=span[0] + ' End')
    ax.fill_betweenx(y=[min(df.min()), max(df.max())], x1=span[1], x2=span[2], color='yellow', alpha=0.2)
    ax.text((span[1] + span[2]) / 2, max(df.max()), span[0], horizontalalignment='center')

# Adding markers for answer sections
questions = json_data['test']
for _, data in questions.items():
    for ts, correct in zip(data['timeStamps'], data['isCorrect']):
        ax.axvline(x=ts, color='blue', linestyle='-')
        ax.text(ts, min(df.min()), f'A', rotation=90, verticalalignment='bottom', horizontalalignment='center')

# Color regions
for i in range(len(activity_spans) - 1):
    ax.axvspan(activity_spans[i][2], activity_spans[i + 1][1], color='lightgrey', alpha=0.5)

plt.tight_layout()

# Save the plot as an image
plt.savefig('brainwave_frequency_bands_with_json_highlighted_and_answered_colored.png')

plt.show()
```
